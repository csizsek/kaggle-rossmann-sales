```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
from pylab import rcParams
%pylab inline
rcParams['figure.figsize'] = 12, 6
from datetime import datetime
import re
```

```python
tr = pd.read_csv('input/train.csv')
tr['Id'] = -1
tr.info()
```

```python
te = pd.read_csv('input/test.csv')
te['Sales'] = -1
te['Customers'] = -1
te.info()
```

```python
s = pd.read_csv('input/store.csv')
s.info()
```

```python
d = pd.concat([tr, te])
```

```python
f = pd.merge(d, s, on='Store', how='left')
```

```python
f[:5]
```

```python
def parse_year(d):
    return int(re.match(r'(\d+)-(\d+)-(\d+)', d).groups()[0])
def parse_month(d):
    return int(re.match(r'(\d+)-(\d+)-(\d+)', d).groups()[1])
def parse_day(d):
    return int(re.match(r'(\d+)-(\d+)-(\d+)', d).groups()[2])
```

```python
f['DateYear'] = f.Date.apply(lambda d: parse_year(d))
f['DateMonth'] = f.Date.apply(lambda d: parse_month(d))
f['DateDay'] = f.Date.apply(lambda d: parse_day(d))
```

```python
def month_in_interval(m, i):
    if i is not None and i != nan and (type(i) == str and (m in i)):
        return 1
    return 0
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for m in months:
    f['PromoInterval_' + m] = f.PromoInterval.apply(lambda i: month_in_interval(m, i))
f[:5].T
```

```python
f.to_csv('tmp/engineered_features.csv', index=False)
```
